```python
import pandas as pd
import sqlite3
place_path = "/kaggle/input/india-ev-market-data/EV Maker by Place.csv"
operationaIPC_path = "/kaggle/input/india-ev-market-data/OperationalPC.csv"
Vehicle_Class_path = "/kaggle/input/india-ev-market-data/Vehicle Class - All.csv"
ev_cat_path = "/kaggle/input/india-ev-market-data/ev_cat_01-24.csv"
ev_sales_by_makers_cat_path = "/kaggle/input/india-ev-market-data/ev_sales_by_makers_and_cat_15-24.csv"
conn = sqlite3.connect('/kaggle/working/Electric_Vehicle_Data.db')
EV_Maker_by_place = pd.read_csv(place_path)
OperationalIPC = pd.read_csv(operationaIPC_path)
Vehicle_Class_All = pd.read_csv(Vehicle_Class_path)
ev_cat_01_24 = pd.read_csv(ev_cat_path)
ev_sales_by_makers_and_cat15_24 = pd.read_csv(ev_sales_by_makers_cat_path)
```

**Adoption rate of electric vehicles over the years**

```python
vecheli_sold = pd.DataFrame(ev_sales_by_makers_and_cat15_24)
column_sums = ev_sales_by_makers_and_cat15_24[['2015','2016','2017','2018','2019','2020','2021','2022','2023','2024']].sum()
column_sums_df = column_sums.reset_index()
column_sums_df.columns = ['Year', 'Total_EV_Sold']
column_sums_df['Growth_Rate'] = column_sums_df['Total_EV_Sold'].pct_change() * 100
column_sums_df['Growth_Rate'] = column_sums_df['Growth_Rate'].fillna(0)
column_sums_df.to_csv('ev_adoption.csv', index=False)
column_sums_df
```

**Geographical Analysis**

```python
state_counts = EV_Maker_by_place['State'].value_counts().reset_index()
state_counts.columns = ['State', 'State_Count']
state_counts.to_csv('geographical_Analysis.csv', index=False)
state_counts
```

**Manufacturer Performance**

```python
manufacturer_performance = ev_sales_by_makers_and_cat15_24.groupby('Maker')[['2015', '2016', '2017', '2018', '2019', '2020', '2021', '2022', '2023', '2024']].sum()
manufacturer_performance['Manufacturer_Sales'] = manufacturer_performance.sum(axis=1)
manufacturer_performance['Manufacturer_Sales'] = manufacturer_performance['Manufacturer_Sales'].fillna(0)
Total_Sales = manufacturer_performance['Manufacturer_Sales'].sum()
manufacturer_performance['Market_Share (%)'] = (manufacturer_performance['Manufacturer_Sales'] / Total_Sales) * 100
manufacturer_performance = manufacturer_performance.reset_index()
manufacturer_performance = manufacturer_performance[['Maker','Manufacturer_Sales', 'Market_Share (%)']]
manufacturer_performance.to_csv('manufacturer_performance.csv', index=False)
print("Total sales across all manufacturers:", Total_Sales)
```
